### World Energy Consumption

Find column descriptions on the link below: → https://www.kaggle.com/datasets/pralabhpoudel/world-energy-consumption

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('World Energy Consumption.csv')
df
```

```python
# Number of Countries Tracked
len(dict(df.country.value_counts()))
```

```python
# Types of energy consumption measured in dataset
np.array(df.columns[::3])
```

```python
# Create sub dataset based on country selection
def SELECT_COUNTRY(country=None, iso=None):
    if country == None:
        return df[df.iso_code == iso.upper()]
    return df[df.country == country.title()]

united_states = SELECT_COUNTRY(iso='USA')
united_states
```

```python
# Sub Dataset based on select energy

# Selectable Energy Types:
# Coal, Biofuel, Carbon, Hydro, Solar, Nuclear, Wind, Oil, Gas, Fossil, Renewable, Other

def SELECT_ENERGY(df, energy):
    col_names = ['iso_code', 'country', 'year'] + [col for col in df.columns if energy in col]
    return df[df.columns.intersection(col_names)]

US_Coal = SELECT_ENERGY(united_states, 'oil')
US_Coal
```

```python
# Visualizing change in same energy type usage (by country)

# Select 4 Countries:
country1 = SELECT_COUNTRY(country='Saudi Arabia') # Change country here by Iso or Country name
country2 = SELECT_COUNTRY(country='Germany')
country3 = SELECT_COUNTRY(country='United Kingdom')
country4 = SELECT_COUNTRY(country='Russia')

# Select energy type (We will be using coal)
en_type = 'oil' # Change energy type here
country1_en = SELECT_ENERGY(country1, en_type) 
country2_en = SELECT_ENERGY(country2, en_type)
country3_en = SELECT_ENERGY(country3, en_type)
country4_en = SELECT_ENERGY(country4, en_type)

# Plotting figure
fig, axs = plt.subplots(2, 2, figsize=(12.5, 12.5))
fig.suptitle(f'{en_type.title()} usage comparison by country, 1900-2020', size=24)

# Country 1
axs[0, 0].plot(country1.year, country1_en.oil_consumption) # Change column name by searching using attribute method
axs[0, 0].set_title(country1.country.values.max(), size=20)
axs[0, 0].set_ylabel('Annual Consumption (in kilo-watt hours)', size=16)
axs[0, 0].set_xlabel('Year', size=12)

# Country 2
axs[0, 1].plot(country2.year, country2_en.oil_consumption, color='r')
axs[0, 1].set_title(country2.country.values.max(), size=20)
axs[0, 1].set_ylabel('Annual Consumption (in kilo-watt hours)', size=16)
axs[0, 1].set_xlabel('Year', size=12)

# Country 3
axs[1, 0].plot(country3.year, country3_en.oil_consumption, color='g')
axs[1, 0].set_title(country3.country.values.max(), size=20)
axs[1, 0].set_ylabel('Annual Consumption (in kilo-watt hours)', size=16)
axs[1, 0].set_xlabel('Year', size=12)

# Country 4
axs[1, 1].plot(country4.year, country4_en.oil_consumption, color='orange')
axs[1, 1].set_title(country4.country.values.max(), size=20)
axs[1, 1].set_ylabel('Annual Consumption (in kilo-watt hours)', size=16)
axs[1, 1].set_xlabel('Year', size=12)

plt.show()
```

```python
# Let's create larger datasets of countries and energies we want
def COMBINE_SETS(*countries:str):
    df = SELECT_COUNTRY(country=countries[0])
    for i in range(1, len(countries)):
        df = df.append(SELECT_COUNTRY(country=countries[i]))
    return df.sort_index()

def COMBINE_EN(country, *en):
    df = SELECT_ENERGY(country, energy=en[0])
    for i in range(1, len(en)):
        data = SELECT_ENERGY(country, energy=en[i])
        for col_name, col_data in data.iteritems():
            df[col_name] = col_data
    return df
```

```python
max_con = COMBINE_SETS('United States', 'Russia', 'China', 'Saudi Arabia', 'India') # Add as many countries here as you wish by country name
max_con
```

```python
sns.jointplot(x=max_con.year, y=max_con.energy_per_capita, hue=max_con.country)
```

```python
sns.jointplot(x=max_con.year, y=max_con.gdp, hue=max_con.country)
```

```python
usa = SELECT_COUNTRY(country='United States')
max_en = COMBINE_EN(usa, 'oil', 'gas', 'solar', 'carbon') # Add any number of energy types for your country dataframe made using SELECT_COUNTRY()
max_en
```

```python
# Visualizing change in different energy usages (by country)

# Plotting figure
fig2, axs2 = plt.subplots(2, 2, figsize=(12.5, 12.5))
fig2.suptitle(f'{max_en.country.values.max()} energy usage comparison by type, 1900-2020', size=24)

# Energy 1
axs2[0, 0].scatter(max_en.year, max_en.oil_consumption)
axs2[0, 0].set_title('Oil', size=18)
axs2[0, 0].set_xlabel('Year', size=14)
axs2[0, 0].set_ylabel('Annual Consumption (in terrawatt-hours)', size=14)

# Energy 2
axs2[0, 1].scatter(max_en.year, max_en.gas_consumption, color='r')
axs2[0, 1].set_title('Gas', size=18)
axs2[0, 1].set_xlabel('Year', size=14)
axs2[0, 1].set_ylabel('Annual Consumption (in terrawatt-hours)', size=14)

# Energy 3
axs2[1, 0].scatter(max_en.year, max_en.solar_consumption, color='g')
axs2[1, 0].set_title('Solar', size=18)
axs2[1, 0].set_xlabel('Year', size=14)
axs2[1, 0].set_ylabel('Annual Consumption (in terrawatt-hours)', size=14)

# Energy 4
axs2[1, 1].scatter(max_en.year, max_en.low_carbon_consumption, color='orange')
axs2[1, 1].set_title('Carbon', size=18)
axs2[1, 1].set_xlabel('Year', size=14)
axs2[1, 1].set_ylabel('Annual Consumption (in terrawatt-hours)', size=14)

plt.show()
```
